```python
import requests
from bs4 import BeautifulSoup
import pandas

r=requests.get("https://www.century21.com/real-estate/california-md/LCMDCALIFORNIA/", headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
c=r.content

soup=BeautifulSoup(c, "html.parser")

all=soup.find_all("div", {"class":"property-card-primary-info"})
```

```python
l=[]
for item in all:
    d={}
    try:
        d["Price"]=item.find("a", {"class":"listing-price"}).text.strip()
    except:
        d["Price"]=None
        
    try:
        d["Address"]=item.find("div", {"class":"property-address"}).text.strip()
    except:
        d["Address"]=None
    
    try:
        d["City"]=item.find("div", {"class":"property-city"}).text.strip()
    except:
        d["City"]=None
    
    try:
        d["Bed"]=item.find("div", {"class":"property-beds"}).find("strong").text.strip()
    except:
        d["Bed"]=None
    
    try:
        d["Bath"]=item.find("div", {"class":"property-baths"}).find("strong").text.strip()
    except:
        d["Bath"]=None
    
    try:
        d["Sqft"]=item.find("div", {"class":"property-sqft"}).find("strong").text.strip()
    except:
        d["Sqft"]=None
    
    l.append(d)
```

```python
df=pandas.DataFrame(l)
```
